```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import GridSearchCV, train_test_split
```

```python
# Dataset de teste - Preço de Casas
dados_casas = {
    "n_quartos": [3, 4, 5, 3, 4, 5],
    "tamanho_lote": [500, 700, 900, 600, 800, 1000],
    "idade_casa": [10, 5, 15, 20, 10, 5],
    "preco": [150000, 200000, 250000, 180000, 220000, 300000]
}

dados = pd.DataFrame(dados_casas)
dados
```

### Como implementar e treinar a árvore de decisão?

```python
# Separando os dados em atributos de entrada (X) e atributo meta (y)
X = dados.drop(columns=['preco'], axis=1)
y = dados['preco']
```

```python
# Carregar biblioteca
from sklearn.tree import plot_tree, DecisionTreeRegressor

# Criar o modelo de árvore de decisão para regressão
model = DecisionTreeRegressor(random_state=33, criterion='absolute_error')

# Treinar o modelo
model.fit(X, y)
```

```python
# Plotar a árvore de decisão
fig, ex = plt.subplots(figsize=(12,8))
plot_tree(model, feature_names=X.columns, filled=True, rounded=True)
plt.show()
```

### Como efetuar novas previsões?

```python
# Predizer novos valores
novos_dados_X = {
    "n_quartos": [4, 3],
    "tamanho_lote": [760, 550],
    "idade_casa": [7, 12]
}

dados_pred= pd.DataFrame(novos_dados_X)
y_pred = model.predict(dados_pred)

# Imprimir valores
dados_pred['preco_previsto'] = y_pred

dados_pred
```
